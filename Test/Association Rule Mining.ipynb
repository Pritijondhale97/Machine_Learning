Association Rule Mining

```python
import pandas as pd
path = r"https://raw.githubusercontent.com/Sindhura-tr/Datasets/refs/heads/main/store_data.csv"
df = pd.read_csv(path,header=None)
df.head()
     
```

```python
records = []
for i in range(df.shape[0]):
  records.append([str(df.values[i,j]) for j in range(df.shape[1])])
records[:50]
     
```

Criteria: Set the minimum values for Confidence, Support, Lift

Confidence = 0.2 support = 7*5/len(df) lift = 3

```python
conf = 0.2
sup = 35/len(df)
lift = 3
     
```

```python
from apyori import apriori
association_rules = apriori(records,min_support=sup,min_confidence=conf,min_lift=lift)
association_rules = list(association_rules)
association_rules[0]
     
```

```python
pair0 = association_rules[0][0]
items0 = []
for i in pair0:
  #print(i)
  items0.append(i)
items0
print("Associaiton rules: " + items0[0] + "----->" + items0[1])
     
```

```python
for item in association_rules:
  pair = item[0]

  items =[i for i in pair]
  print("Association Rule: " + items[0]  + "-------->" + items[1])

  sup = item[1]
  print("Support: "+str(item[1]))

  conf = str(item[2][0][2])
  lift = str(item[2][0][3])
  print("Confidence : "+ conf)
  print("Lift : "+lift)
  print("--------------------------------")

     
```

```python
rules = []
supports = []
confs = []
lifts =[]

for item in association_rules:
  pair = item[0]

  items =[i for i in pair]
  ru = (items[0]  + "-->" + items[1])
  rules.append(ru)


  sup = item[1]

  supports.append(sup)

  conf = (item[2][0][2])
  lift = (item[2][0][3])

  confs.append(conf)
  lifts.append(lift)
     
```

```python
rules
     
```

```python
supports
```

Create a dictionary to save the associations formed

```python

dct = {
    "Rules":rules,
    "Support":supports,
    "Confidence":confs,
    "Lift":lifts
}
     
```

```python
from typing_extensions import final
final_associations = pd.DataFrame(dct)
final_associations
     
```

Save the results to output file

```python
final_associations.to_csv("Associations_Ready.csv",index=False)
     
```
